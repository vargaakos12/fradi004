1\. Első rendelés

```sql
SELECT MIN(REND_DATUM) AS 'Első rendelés'
FROM Rendeles
```

2\. Utolsó rendelés

```sql
SELECT MAX(REND_DATUM) AS 'Utolsó rendelés'
FROM Rendeles
```

3\. Null-Nem Null sorok aránya

```sql
SELECT COUNT(*) AS 'Összes',
        COUNT(*)-COUNT(REND_DATUM) AS 'Null',
        COUNT(REND_DATUM) AS 'Nem Null'
FROM Rendeles
```

4\. Rendelések száma évente

```sql
SELECT YEAR(REND_DATUM) AS 'Év',
        COUNT(*) AS 'Rendelések száma'
FROM Rendeles
GROUP BY YEAR(REND_DATUM)
ORDER BY 2 DESC
```

5. Bankártyával fizetett és szállítási címet megadott rendelések

```sql
SELECT FIZ_MOD AS 'Fizetési mód',
        SZALL_CIM AS 'Szállítási cím'
FROM Rendeles
WHERE SZALL_CIM IS NOT NULL AND FIZ_MOD LIKE 'Bankkártya'
```
